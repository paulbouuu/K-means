```python
import numpy as np
import matplotlib.pyplot as plt

from k_means import K_means
from utils import generate_2D_clusters
```

```python
k = 6
domain = [-10,10]

X = generate_2D_clusters(k, points_per_cluster = 200, domain = domain)

plt.scatter(X[:,0], X[:,1], s=5, alpha=0.5)
plt.show()
```

```python
# ------------------------------------------
# Parameters for k-means algorithm
iterations = 20
plotting = False
create_gif = True
# ------------------------------------------


k_means = K_means(k = k, domain = domain)

k_means.init_k_means_algorithm(X)


for _ in range(iterations):
    
    k_means.step()
    
    k_means.plot(plotting = plotting)


if create_gif:
    from utils import gif
    gif(duration = 500)
```
